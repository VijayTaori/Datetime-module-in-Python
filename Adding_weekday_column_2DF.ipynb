```python
import pandas as pd
import numpy as np
import calendar
import datetime
df_usd2eur = pd.read_csv('/content/USD2EUR_all.csv')
df_usd2eur['Date'] = pd.to_datetime(df_usd2eur['Date'],dayfirst=True)
def find_weekday(y):
  day = calendar.day_name[y.weekday()]
  return day
df_usd2eur['Weekday'] = df_usd2eur['Date'].map(find_weekday)
df_usd2eur
```
