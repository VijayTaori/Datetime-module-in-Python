```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df_open = pd.read_csv('/content/crypto_et_open_ut.csv')
print(df_open.columns)
```

```python
df_open['Open Datetime'] = pd.to_datetime(df_open['Open Datetime'],dayfirst=True)
df_open['Open Date_x'] = pd.to_datetime(df_open['Open Date_x'],dayfirst=True)
```

```python
import time
```

```python
def convert2UnixTimestamp(date_time):
  unix_timestamp = time.mktime(date_time.timetuple())
  return unix_timestamp
```

```python
df_open
```

```python
df_open['open_unix_timestamp2'] = df_open['Open Datetime'].map(convert2UnixTimestamp)
```
